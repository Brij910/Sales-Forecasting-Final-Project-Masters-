## Arima Model

```python
Tr_X = train2["sales"].values
Te_X = val2["sales"].values
sales_history = [x for x in Tr_X]
arima_te_prediction = list()
from statsmodels.tsa.arima.model import ARIMA
for t in range(len(Te_X)):
    model = ARIMA(sales_history, order=(5,1,0))
    model_fit = model.fit()
    output = model_fit.forecast()
    yhat = output[0]
    arima_te_prediction.append(yhat)
    obs = Te_X[t]
    sales_history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
```

```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(train2["sales"][0:100])
plt.savefig('autocorrelation_plot_Arima_plot_2.png')

plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rmse2 = sqrt(mean_squared_error(Te_X, arima_te_prediction))
print('Test RMSE: %.3f' % rmse2)
plt.plot(Te_X[30:60])
plt.plot(arima_te_prediction[30:60], color='red')
plt.show()
```

```python
plt.plot(val2["sales"].values[30:60], label= "Actual Sales")
plt.plot(arima_te_prediction[30:60], color='red', label='Predicted Value')
plt.title("Arima Model Plot")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.legend(loc = "upper right")
plt.savefig('Arima_plot_2.png')
plt.show()
```

## LSTM Model

```python
# Considering time step of 10 
# Model will predict 11th sale value by looking at previous 10 results.

time=10
def create_dataset(dataset, time_step):
    train_feature, train_sales = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step), 0]    
        train_feature.append(a)
        train_sales.append(dataset[i + time_step, 0])
    return np.array(train_feature), np.array(train_sales)
```

```python
new_data_lstm = pd.concat([train2, val2])
```

```python
new_data_lstm
```

```python
new_data_lstm = new_data_lstm.drop('onpromotion', axis = 1)
new_data_lstm = new_data_lstm.drop('dcoilwtico', axis = 1)
new_data_lstm = new_data_lstm.drop('transactions', axis = 1)
```

```python
new_data_lstm
```

```python
sc=MinMaxScaler(feature_range=(0,1))
new_data_lstm=sc.fit_transform(np.array(new_data_lstm).reshape(-1,1))
```

```python
new_data_lstm
```

```python
new_data_lstm_length_train = int(len(new_data_lstm)*0.7)
new_data_lstm_length_test = len(new_data_lstm)-new_data_lstm_length_train
```

```python
new_data_lstm_train = new_data_lstm[:new_data_lstm_length_train,:]
new_data_lstm_test = new_data_lstm[new_data_lstm_length_train:,:1]
```

```python
new_data_lstm_test
```

```python
X_train, y_train = create_dataset(new_data_lstm_train, time)
X_test, ytest = create_dataset(new_data_lstm_test, time)
```

```python
X_train.shape
```

```python
X_train =X_train.reshape(X_train.shape[0],X_train.shape[1],1)
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1],1)
```

```python
model=Sequential()

# We have used relu activation function
# Input shape will be (10,1) because we have taken time-step of 10
model.add(LSTM(50, activation='relu',return_sequences=True,input_shape=(10,1)))
model.add(LSTM(75,return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))

# For the compilation of model we have used adam optimizer with learning rate of 0.001
# FOr the evaluation we have calculated mse values 
model.compile(loss="mse", optimizer=Adam(learning_rate=0.001), metrics=['mse'])
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,validation_data=(X_test,ytest),epochs=200,batch_size=64,verbose=1)
```

```python
plt.plot(model.history.history['mse'])
plt.plot(model.history.history['val_mse'])
plt.xlabel("Epochs")
plt.ylabel("MSE")
plt.legend(['Train', 'Validation'])
plt.title("Mean Squared Error - LSTM with Dates")
plt.savefig("MSE_LSTM_Date_2.png")
plt.show()
```

```python
train_pred=model.predict(X_train)
test_pred=model.predict(X_test)
train_pred=sc.inverse_transform(train_pred)
test_pred=sc.inverse_transform(test_pred)
```
