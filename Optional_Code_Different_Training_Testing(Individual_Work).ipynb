## Arima Model 

```python
X = Total_sales2["sales"].values
size = int(len(X) * 0.66)
train2, test2 = X[0:size], X[size:len(X)]
history2 = [x for x in train2]
predictions2 = list()
from statsmodels.tsa.arima.model import ARIMA
for t in range(len(test2)):
    model = ARIMA(history2, order=(5,1,0))
    model_fit = model.fit()
    output = model_fit.forecast()
    yhat = output[0]
    predictions2.append(yhat)
    obs = test2[t]
    history2.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
from sklearn.metrics import mean_squared_error
from math import sqrt
rmse2 = sqrt(mean_squared_error(test2, predictions2))
print('Test RMSE: %.3f' % rmse2)
plt.plot(test2[0:30])
plt.plot(predictions2[0:30], color='red')
plt.show()
```

```python
plt.plot(test2[30:60], label= "Actual Sales")
plt.plot(predictions2[30:60], color='red', label='Predicted Value')
plt.title("Arima Model Plot")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.legend(loc = "upper right")
plt.savefig('Arima_plot.png')
plt.show()
```

## LSTM with Family Feature

```python
holidays=pd.read_csv('holidays_events.csv')
oil=pd.read_csv('oil.csv')
stores=pd.read_csv('stores.csv')
test=pd.read_csv('test.csv')
train=pd.read_csv('train.csv')
transactions=pd.read_csv('transactions.csv')
```

```python
holidays['date'] = pd.to_datetime(holidays['date'])
oil['date'] = pd.to_datetime(oil['date'])
train['date'] = pd.to_datetime(train['date'])
test['date'] = pd.to_datetime(test['date'])
transactions['date'] = pd.to_datetime(transactions['date'])
```

```python
train.shape
```

```python
from sklearn.model_selection import train_test_split
test_new_data = train.iloc[2500001:,:]
train = train.iloc[0:2500000,:]
```

```python
train["family"].nunique(dropna=True)
test_new_data["family"].nunique(dropna=True)
```

```python
test_new_data
```

```python
training_data = train.copy().drop(['onpromotion'], axis=1)
test_new_data_dp = test_new_data.copy().drop(['onpromotion'], axis=1)
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import MinMaxScaler
ordinal_encoder = OrdinalEncoder(dtype=int)
training_data[['family']] = ordinal_encoder.fit_transform(training_data[['family']])
test_new_data_dp[['family']] = ordinal_encoder.fit_transform(test_new_data_dp[['family']])
```

```python
pivoted_train = training_data.pivot(index=['date'], columns=['store_nbr', 'family'], values='sales')
pivoted_train.head()
```

```python
pivoted_train[1][0]
```

```python
n_o_days_train=training_data["date"].nunique(dropna = False) 
print('number of day train:',n_o_days_train)

n_o_stores_train=training_data["store_nbr"].nunique(dropna = False) 
print('number of stores train:',n_o_stores_train)

n_o_families_train=training_data["family"].nunique(dropna = False) 
print('number of family/type of prod train:',n_o_families_train)
```

```python
train_samples = int(n_o_days_train*0.90)
train_samples
```

```python
train_samples_df = pivoted_train[:train_samples]
train_samples_df
```

```python
valid_samples_df = pivoted_train[train_samples:]
valid_samples_df = valid_samples_df.dropna()
```

```python
minmax = MinMaxScaler()
minmax.fit(train_samples_df)

scaled_train_samples = minmax.transform(train_samples_df)
scaled_val_samples = minmax.transform(valid_samples_df)
```

```python
scaled_train_samples[10:]
```

```python
def split_series(series, n_past, n_future):
    X, y = list(), list()
    for window_start in range(len(series)):
        past_end = window_start + n_past
        future_end = past_end + n_future
        if future_end > len(series):
            break
            
        past, future = series[window_start:past_end,:], series[past_end:future_end,:]
        X.append(past)
        y.append(future)
    
    return np.array(X), np.array(y)

n_past =5
n_future = 5
n_features = n_o_stores_train * n_o_families_train # num of features
```

```python
X_train, y_train = split_series(scaled_train_samples, n_past, n_future)
X_val, y_val = split_series(scaled_val_samples, n_past, n_future)
```

```python
print('X_train.shape',X_train.shape)
print('y_train.shape',y_train.shape)
print('X_val.shape',X_val.shape)
print('y_val.shape',y_val.shape)
```

```python
model = Sequential()

model.add(LSTM(units=256, return_sequences=True,input_shape=[n_past, n_features]))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(LSTM(units=128, return_sequences=True))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(TimeDistributed(Dense(n_features)))

model.compile(loss="mse", optimizer=Adam(learning_rate=0.001), metrics=['mse'])
```

```python
model.summary()
```

```python
model_history = model.fit(X_train, y_train, 
                          validation_data=(X_val, y_val),
                          epochs = 100,
                          batch_size=512,
                          shuffle=True)
```

```python
plt.plot(model.history.history['mse'])
plt.plot(model.history.history['val_mse'])
plt.xlabel("Epochs")
plt.ylabel("MSE")
plt.title("Mean Squared Error - LSTM with Family")
plt.legend(['Train', 'Validation'])
plt.savefig("MSE_LSTM_Family.png")
plt.show()
```

## LSTM with Considering Dates and Trend

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense, Dropout
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
holidays=pd.read_csv('holidays_events.csv')
oil=pd.read_csv('oil.csv')
stores=pd.read_csv('stores.csv')
test=pd.read_csv('test.csv')
train=pd.read_csv('train.csv')
transactions=pd.read_csv('transactions.csv')
```

```python
holidays['date'] = pd.to_datetime(holidays['date'])
oil['date'] = pd.to_datetime(oil['date'])
train['date'] = pd.to_datetime(train['date'])
test['date'] = pd.to_datetime(test['date'])
transactions['date'] = pd.to_datetime(transactions['date'])
```

```python
train
```

```python
import tensorflow as tf
import keras
from sklearn.preprocessing import MinMaxScaler
```

```python
sc=MinMaxScaler(feature_range=(0,1))
train=sc.fit_transform(np.array(train).reshape(-1,1))
```

```python
train.shape
```

```python
train_length = int(len(train)*0.7)
test_length = len(train)-train_length
```

```python
train_set = train[:train_length,:]
test_set = train[train_length:,:1]
```

```python
test_set
```

```python
time=10
def create_dataset(dataset, time_step):
    dataX, dataY = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step), 0]    
        dataX.append(a)
        dataY.append(dataset[i + time_step, 0])
    return np.array(dataX), np.array(dataY)
```

```python
time_step = 10
X_train, y_train = create_dataset(train_set, time)
X_test, ytest = create_dataset(test_set, time)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
X_train =X_train.reshape(X_train.shape[0],X_train.shape[1],1)
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1],1)
```

```python
X_train
```

```python
model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(10,1)))
model.add(LSTM(75,return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss="mse", optimizer=Adam(learning_rate=0.001), metrics=['mse'])
```

```python

model.summary()
```

```python
model.fit(X_train,y_train,validation_data=(X_test,ytest),epochs=200,batch_size=64,verbose=1)
```

```python
plt.plot(model.history.history['mse'])
plt.plot(model.history.history['val_mse'])
plt.xlabel("Epochs")
plt.ylabel("MSE")
plt.legend(['Train', 'Validation'])
plt.title("Mean Squared Error - LSTM with Dates")
plt.savefig("MSE_LSTM_Date.png")
plt.show()
```
